Exercise 1: Pizza Toppings ☑️
Write a loop that prompts the user to enter a series of pizza toppings until they enter a 'quit' value. As they enter each topping, print a message saying you’ll add that topping to their pizza.

```python
print('Enter the topping you want to add on your pizza.')
print('Type quit to exit')

while True:
  topping = str(input())
  print(topping + 'will be added on your pizza.')

  if topping == 'quit':
    break

  else:
      print('You have added the following topping:\n' +str(topping).title())  
```

Exercise 2: Movie Tickets: ☑️
A movie theater charges different ticket prices depending on a person’s age. If a person is under the age of 3, the ticket is free; if

they are between 3 and 12, the ticket is $10; and if they are over age 12, the ticket is $15. Write a loop in which you ask users their age, and then tell them the cost of their movie ticket

```python
print('Enter your age')
print('Type quit to exit')

age=0

while True:
  age = input()
  if age == 'quit':
    break

  else:
     age = int(age)
     if age < 3 :
       print('The theater ticket is free for you.') 

     elif 3 <= age <= 12:
       print('The theater ticket price is $10.')

     elif age > 12:
       print('The theater ticket price is $15')    
```


Exercise 4: Deli ☑️
Make a list called sandwich_orders and fill it with the names of various sandwiches. Then make an empty list called finished_sandwiches.

Loop through the list of sandwich orders and print a message for each order, such as I made your tuna sandwich. As each sandwich is made, move it to the list of finished sandwiches. After all the sandwiches have been made, print a message listing each sandwich that was made.

```python
sandwich_orders = ['tuna', 'turkey', 'grilled chicken', 'beef', 'veggie', 'pastrami', 'pastrami', 'pastrami']
finished_sandwiches = []

while sandwich_orders:
  current_sandwich = sandwich_orders.pop()
  print('Your ' + current_sandwich + ' sandwich is ready')

  finished_sandwiches.append(current_sandwich)

print('\n')
print('Sandwich Orders:')

for sandwich in finished_sandwiches:
  print(sandwich, 'Sandwich')
```

Exercise 5: No Pastrami ☑️
Using the list sandwich_orders from Exercise 7-8, make sure the sandwich 'pastrami' appears in the list at least three times. Add code near the beginning of your program to print a message saying the deli has run out of pastrami, and then use a while loop to remove all occurrences of 'pastrami' from sandwich_orders. Make sure no pastrami sandwiches end up in finished_sandwiches.

```python
sandwich_orders = ['tuna', 'turkey', 'grilled chicken', 'beef', 'veggie', 'pastrami', 'pastrami', 'pastrami']
finished_sandwiches = []

print('The deli has run out of pastrami')

for sandwich in sandwich_orders:
  if sandwich == 'pastrami':
    sandwich_orders.remove(sandwich)

while sandwich_orders:
  current_sandwich = sandwich_orders.pop()
  print('I made your ' + current_sandwich + 'sandwich.')
  finished_sandwiches.append(current_sandwich)

print('\n')
print('Sandwich Orders:')

for sandwich in finished_sandwiches:
  print(sandwich, 'Sandwich')

print(sandwich_orders)  
```
